```python
import numpy as np

for f in ["data_eeg/neural_data_label1_1.npy", "data_eeg/neural_data_label2_1.npy", "data_eeg/neural_data_label3_1.npy"]:
    x = np.load(f)
    print(f, "shape:", x.shape, "dtype:", x.dtype, "min/max:", x.min(), x.max())
```

```python
label_to_prompt = {
    "label1": "a person walks forward steadily",
    "label2": "a person squats down and stands up",
    "label3": "a person turns around in place"
}

def filename_to_label(path: str):
    for k in ["label1", "label2", "label3"]:
        if k in path:
            return k
    raise ValueError("Cannot parse label from filename: " + path)
```

```python
def eeg_to_motion(eeg_path, out_example="assets/eeg_example.txt", length=120):
    label = filename_to_label(eeg_path)
    prompt = label_to_prompt[label]
    Path(out_example).write_text(f"{length} {prompt}\n", encoding="utf-8")
```
